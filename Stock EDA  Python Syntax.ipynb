# Stock Price Exploratory Data Analysis

Completing an exploratory data analysis (EDA) of the price action of several large-cap companies during a specific time period. 

```python
#importing packages

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#importing data frame

df = pd.read_csv(r"C:\Users\togarro\Downloads\Stocks.csv")
```

```python
# converting date column to ISO 8601 format (YYYY-MM-DD)

df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.shape
```

The dataset contains 248 rows and 8 columns.

```python
# Date metrics

most_recent = df['Date'].max()
least_recent = df['Date'].min()

range = most_recent - least_recent

print(most_recent)
print(least_recent)
print(range)
```

The dataset contains price data of the stock market for a total of 87 days beginning in February 7, 2023 and ending in May 5, 2023.

```python
# Stock tickers

stock_symbols = df['Ticker'].unique()

print(stock_symbols)
```

The dataset contains the price action data of four companies, Apple - AAPL, Microsoft - MFST, Netflix - NFLX, and Google - GOOG. All of the companies are large cap companies, meaning they have at least a $10 Billion valuation

```python
# Statistical Metrics AAPL

AAPL = df.query('Ticker == "AAPL"')

AAPL['Close'].describe()
```

```python
# Creating subplot
fig, axs = plt.subplots(nrows =1 , ncols =2 , figsize = (15,5))

# AAPL Histogram
AAPL['Close'].plot(kind = 'hist', bins = 20, ax = axs[0], title = 'AAPL Histogram')

# AAPL Boxplot 
AAPL['Close'].plot(kind = 'box', ax = axs[1], title = 'AAPL Boxplot',ylabel = 'Price', vert = False)
```

### AAPL Distribution Analysis

The histogram of AAPL price action data is trimodal. Based on the boxplot, there is a positive skew of the AAPL price action data. 

```python
# Statistical Metrics MSFT

MSFT = df.query('Ticker == "MSFT"')

MSFT['Close'].describe()
```

```python
# Creating subplot 
fig, axs1 = plt.subplots(nrows =1 , ncols =2 , figsize = (15,5))

# MFST Histogram
MSFT['Close'].plot(kind = 'hist', bins = 30, ax = axs1[0], title = 'MSFT Histogram')

# MFST Boxplot
MSFT['Close'].plot(kind = 'box', ax = axs1[1], title = 'MSFT Boxplot', ylabel = 'Price', vert = False)
```

### MSFT Distribution Analysis 

The histogram of MSFT price action data is indicative a multimodal distribution. Based on the boxplot there is a left skew in the price action data of MSFT.
